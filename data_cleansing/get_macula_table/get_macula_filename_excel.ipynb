这个文件提取了`macula_grading_20240805.xlsx`的`after`作为`macula_filename`列, `grading macular+optic disc`作为`label`列, 并根据文件名进行解析为pid, eye, date, time

```python
import pandas as pd
```

```python
df = pd.read_excel(
    "/home/xujialiu/mnt-4T-xujialiu/my_lightning_model/tabular_data/macula_grading_20240805.xlsx"
)
```

```python
df
```

```python
df_clean = pd.DataFrame()
```

```python
df_clean["label"] = df["grading macular+optic disc"]
df_clean["macula_filename"] = df["after"]
```

```python
df_clean
```

```python
df_clean["pid"] = df_clean.macula_filename.str.split("_").str[0]
df_clean["eye"] = df_clean.macula_filename.str.split("_").str[1]
df_clean["date"] = df_clean.macula_filename.str.split("_").str[2]
df_clean["time"] = df_clean.macula_filename.str.split("_").str[3]

df_clean["date"] = pd.to_datetime(df_clean.date, format=r"%Y%m%d")
pd.to_datetime(df_clean["time"], format=r"%H%M%S")
df_clean["time"] = df_clean["time"].apply(lambda x: ":".join([x[:2], x[2:4], x[4:]]))
df_clean["date"] = df_clean["date"].astype(str)
```

```python
df_clean
```

```python
df_clean = df_clean.dropna(how="any")
```

```python
df_clean.to_excel("macula_table.xlsx")
```
